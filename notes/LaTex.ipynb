# 2.2 命令
用户命令可以分为两种：
1. 反斜杠符号后面跟由若干个英文字母组成的命令名，区分大小写，以空格、数字（个别命令如\\fontdimen等除外）或非字母符号作为结束标志。这种命令后的所有空格都被忽略。如果想保留命令之后的一个空格可以用`{} `或`\ `。因为系统将`{}`视为高和宽为零的非字母符号。  
2. 反斜杠符号后面跟由一个非字母符号的命令名。这种命令无需任何字符或空格作为结束标志，因为系统只认反斜杠后的第一个符号为命令名。这种命令系统会保留命令后第一个空格。  

编写命令：  
用于编写文类或者宏包文件。如`\RequirePackage`，通常是较长的大小写混合字母串组成。  
内部命令：  
系统内部使用的命令。其中的一部分在命令名中含有`@`，又被称为核心命令。如`\@author`，系统不会将源文件中`@`当做字母，而是符号，而在系统本身和文类或宏包中则当做字母，所以这类命令都不能直接用于源文件中。  
自定义命令：  
`\newcommand{新命令}[参数数量][默认值（只能有一个，会使得新命令的第一个参数变为可选参数，新命令可选参数只能有一个）]{定义内容}`。  
`\newcommand*{新命令}[参数数量][默认值]{定义内容}`。命令中各种参数不可以含有`\par`或者空行，即每一个参数的内容不能超过一个段落。  
前者称为长命令，后者称为短命令。通常使用后者，因为出错时系统可以定位到出错位置。  
`%`：源文件中改符号及其右侧的所有字符，在编译时都将被忽略，用在行末防止文本或者命令换行时被插入多余的空格。  

# 2.5 模式
模式也称为排版模式，它是处理源文件的方式。
LaTex的绘图模式主要用于绘图环境picture,实际上是一种受限的左右模式。  
文本模式：左右模式和段落模式，主要用于文本排版，很多系统命令只能在这两个模式里使用。  
## 2.5.1 左右模式
左右模式（受限水平模式）：系统将其中的内容看作是由单词、标点符号和空白组成的一串字符，无论这串字符有多长，系统都将从左到右顺序排版而不换行；如果超出设定的行宽，系统将提示溢出：`Overfull \hbox`
## 2.5.2 段落模式
段落模式：系统将其中的内容看作是由单词、标点符号和空白组成的一串字符，系统自动对这串字符进行分行、分段和分页处理。段落模式是最常用的排版模式，常规文本的排版就是采用这一模式。段落模式是系统默认的排版模式。  
LaTex内部将段落模式分解为水平模式和垂直模式。  
## 2.5.3 数学模式
很多数学式是二维结构，而编排数学式只能采用字符和命令结合的一维形式。所有的数学符号命令和数学式排版命令都必须在数学模式中使用。某些在文本模式中使用的命令也可应用于数学模式。系统将其中的字母都认为是数学符号，把它们的字体由直立形状转换为斜体形状，并且忽略所有符号之间的空白。

### 2.11.4 居中环境和命令
`\begin{center} \end{center}`和`\centering`功能是相同的，主要区别是前者所居中的文本与上下文之间距离是当前的文本行距，后者有约一行的空白。  
如果只是需要将一行文本，或者某个插图、表格居中，也可使用行居中命令`\centerline{对象}`，其居中文本与上下文间距也等于当前文本行距，但是采用左右模式排版，即使用换行命令`\\`或`\newline`也无效。 
### 2.11.5 左对齐环境和命令
`\begin{flushleft} \end{flushleft}`和`\raggedright`，前者文本与上下文间距等于行距，后者则明显大于行距。行左对齐是`leftline{对象}`，文本与上下文间距也是行距，不过采用左右模式排版，文本超过一行则多余部分凸入右边空。
### 2.11.6 右对齐环境和命令
`\begin{flushright} \end{flushright}`和`\raggedleft`，`rightline{对象}`，其余同上。
**Ps**:系统提供的指令都关闭了断词功能，可能导致每行文本宽度相差很大，为了支持断词功能，可以使用ragged2e对齐宏包。

### 2.11.8 抄录环境和命令
`\begin{verbatim} \end{verbatim}`或`\begin{verbatim*} \end{verbatim*}`可将其中的文本按照其原有的书写格式和字符形态，包括专用符号、空格以及换行位置等，逐字逐句逐空格地抄写记录下来，默认字体为等宽体。  
##### 抄录命令\verb
`verb符号 文本 符号` `\verb*符号 文本 符号`。只适用于抄录一行以内的英文文本，否则将凸出文本行进入右边空。这里的符号可以是没有出现在文本中的，且不是\*和空格的任何符号或数字，通常采用无方向的双引号"或定界符|等没有左右之分的符号。  

抄录命令与抄录环境的区别：
1. 抄录命令及其参数和界限符号都必须在同一个文本编辑行中，否则系统将提示出错。
2. 抄录命令不能对英文进行自动换行，所以只能抄录一行以内的英文文本；而对中文或中英文混排文本却可以自动换行，故能抄录多行以中文为主的文本。抄录环境也不能对英文自动换行，但对以中文为主的文本却可以。
3. 抄录环境和命令不能在其他任何系统命令中使用（除非使用cprotect宏包提供的\cprotect命令），但它们可以在其他环境中使用。
4. 抄录环境中的文本与文本行的左侧边对齐；抄录命令中的文本若处在一段之首，将被缩进处理。  
抄录环境中的文本不应有\，否则改变抄录字体后可能改为半角双引号，因为系统默认字体中只有等宽体有反斜杠号。

**fancyvrb**宏包提供了各种各样的抄录样式。  
**listings**宏包提供了常用于各种编程语言源程序的抄录。

### 2.11.10 宽松环境和命令
为保证排版效果，系统对字词间距、断词、和换行都有严格要求。有时，过于严格会造成断词、多出的字母凸出文本行右端。可以使用宽松环境`\begin{sloppypar} \end{sloppypar}`。  
当有文本行宽度超过设定值时，会有Overfull \hbox警告，当单词间距过宽时，会有Underfull \hbox警告。  
声明形式宽松命令`\sloppy`。宽松环境可用于段落之中，宽松命令只能用于段落之间。

### 2.11.13 自定义环境
每次定义一个新环境，新环境定义会自动将其分解为两个新定义命令。  
`\newenvironment{新环境}[参数数量][默认值]{开始定义}{结束定义}` = `\newcommand{\新环境}[参数数量]{开始定义} + \newcommand{\end新环境}{结束定义}`  
`\ignorespaces`通知系统忽略所有空格直到读入非空格符号为止。  
`\ignorespaceafterend`相当于在环境结束命令`\end`后追加一个`ignorespaces`。

### 2.12加减乘除
需要对两个或两个以上的数值或长度作算术运算时，可调用calc算数宏包。其中的表达式中不能直接用小数进行乘除，应使用下述两条命令之一对小数进行整型转换：`\real{十进制数}`, `\ratio{长度表达式}{长度表达式}`。第一个命令直接对非整数十进制数进行转换，第二个命令用两个长度表达式之比进行转换。  
整数表达式中，小数经过转换可进行乘除运算，运算过程中保留小数的作用，最终结果忽略小数部分。  
长度表达式中，小数经过转换可进行乘除运算，最终结果也将保留小数部分。  
**长度运算的注意事项**：
1. 长度表达式中的各项都必须是长度，其长度单位可以各不相同。  
2. 乘法和除法中，被乘数和被除数必须是长度，乘数和除数必须是整数或是经过整形转换的实数。  
3. 弹性长度在作乘除时，其3个组成都将分别被乘除。

# 3.1字体
LaTex中，字体分为文本字体和数学字体两种，通常文本字体只能在文本模式中使用，数学字体只能在数学模式中使用。  
传统的自体选择方式是用指定字体的属性来确定的，作者无需了解字体的名称和访问的路径，但只限于CTex系统内的字体。出现XeTex后，字体可以直接使用字体名或字体文件名来确定，且适用于计算机内任何TrueType或OpenType字体，只要知道存储路径即可。  
## 字体属性
每种文本字体有五个属性。见[字体属性](./字体属性.jpg)  
字体的每个属性都有多个可选值，只有分别确定每个属性的可选值，字体才能最终被确定。  
使用LaTex时会遇到两种编码。输入编码是通过键盘或者复制粘贴输入到文本编辑器的字符编码，主要WinEdt中主要有ANSI和UTF-8。输出编码是编译输出到pdf中的编码。  
字体文件，俗称字库，是某一类字型和符号的显示描述文件；编码就是在字体文件中查找字符的索引表。  
文本字体的默认编码是OT1，OT1编码的字符只有128个，包括英文字母、阿拉伯数字和标点符号。如需用到西欧的变音符号，只能使用符号命令生成。T1编码的字符有256个，增加了128个西欧字符。  
每种编码都有其默认字族。CTeX附带的文本字体有上千个，被分为三类：
* 罗马体字族
* 等线体字族
* 等宽体字族

根据笔画的粗细宽窄，字体可分为十多种序列，LaTeX采用了其中常规和粗宽两种序列，默认为常规序列。  
字体按外表形态可分为多种形状，LaTex采用了其中直立、斜体、倾斜和小型大写四种，默认为直立。  
字体的尺寸，LaTeX预定为5~25pt。book等3个标准文类对常规字体尺寸的默认值均设定为10pt。 
## 字体设置命令
[字体设置命令及其样式](./字体设置命令,jpg)  
常规字体就是不使用任何字体设置命令所生成的文本字体。  
`\normalfont` `\textnormal`用于恢复使用常规字体，但不改变字体尺寸。在LaTex中是OT1编码、罗马体字族、常规序列和直立形状。  
表中所有简化形式的字体设置命令都由book文类提供，是针对常规字体的。如`\bf`的定义是`\normalfont\bfseries`。
如果希望明确地界定声明形式或简化形式的字体设置命令的作用范围，也可将其命令名作为环境名。  
## 三种字族的视觉效果
罗马体字和等宽体字易于小号字体的识别，但大号字体看起来瘦弱无力。  
罗马体和等宽体的明显区别在于后者每一个字的宽度基本相同。罗马体笔画粗细随横竖变化，等宽体保持不变，等宽体又称为打字机体，因早年的机械英文打字机都使用这种字体。  
等线体没有字头字脚（笔画开始和结束处的装饰衬线），大字体显示时，显得丰满有力，但小字体显示时，容日引起视觉疲劳。  
通常论文的正文使用罗马体，专有名词或者程序命令选用等宽体，章节标题使用等线体。  
### 各种字体形状的区别
斜体形状命令`\textit`与倾斜命令`\textsl`的区别在于前者改变了原字体，后者不改变原字体而只是将其右倾斜一个角度。斜体字体常用于数学式中变量的名称，而倾斜体常用于正文和参考文献中的书刊名称。  
字体的倾斜程度由基本命令`\fontdimen1`设定。  
斜体形状(Italic Shape)直译是意大利体形状，意大利体是一类字体的统称，有斜体形状、直立形状和小型大写形状。  
小型大写形状命令介意将小写字母转换为较当前字体尺寸小一些的大写字母。常用于全部大写的标题、语句或段落。  
强调命令`\emph`或`\em`具有可变字体形状的功能，即可将直立形字体变为斜体形或倾斜形，也可反过来。  
### 复合字体设置命令
不是任何的字体设置命令组合都可以得到相应的复合字体，很多字族没有小型大写形状字体，也没有粗宽字体。编译时，系统没有找到所设定的复合字体，将给出字体信息或警告信息，并自行采用它认为最接近的字体加以替代。  
### 位图字体和向量字体
位图字体主要用于由LaTex编译源文件后所产生的DVI格式文件。位图字体的扩展名是.pk。如使用PDFLaTeX编译源文件直接生成PDF格式文件，或把DVI文件转换为PS或PDF文件时，将改用向量字体。  
向量字体主要有3种：  
1. TrueType，使用二次贝塞尔曲线来描述字形，文件扩展名是.ttf。  
2. Type1，使用三次贝塞尔曲线来描述字形，故比TrueType字体更加精美，文件扩展名是.pfb。  
3. OpenType，集中了前两者的优点并可兼容前两者，扩展名是.otf。  

三种向量字体都可以跨系统使用。  
## 字体尺寸命令
字体尺寸命令都是声明形式，改变其后包括数学模式中的所有字体，直到当前环境或组合结束。  
[字体尺寸命令](./字体尺寸命令.jpg)  
各种字体尺寸命令所对应的字体大小，就是字体的点数，与所用文类的常规字体尺寸有关。  
如希望明确地界定声明形式的字体尺寸命令的作用范围，可以将其命令名作为环境名，组成字体尺寸环境。  
## 局部字体修改  
有时希望用一种截然不同的字体来凸显个别字词，而又不改变系统原来的字体设置，可以使用系统提供的自定义字体设置命令：`\DeclareFixedFont{命令}{编码}{字族}{序列}{形状}{尺寸}`，命令参数不能与已有命令重名，否则将覆盖其定义。  
ptm字族是Times New Roman  
## 文本数字
常用字体中，所有阿拉伯数字的高度都是相等的，故称等高数字。文本数字也称非等高数字、小写数字或旧体数字。  
文本文字清新活泼，无拘无束。常用于电话号码、门牌号码，如用于科普读物的页码，可以使用下列重新定义命令`recommend{\thepage}{\oldstylenums{\the\value{page}}}`。如使用页码设置明亮`\pagenumbering{arabic}`，应将其置于上述命令前，否则将恢复原始定义。  
## 字体定义文件
字体定义文件的作用是以定义命令的形式告诉LaTeX，该字族都有些什么属性的字体，对应的字体文件名是什么。这些字体定义命令通常1只用于字体定义文件中，但也可用在文类或宏保文件中，甚至在源文件的导言中。  
每种字族的字体定义文件名可由下式确定：*编码+字族+.fd=字体文件定义名*   
编码名在各种字体命令中都是大写，但作为文件名的一部分时应该小写。每个字体定义文件都有一条字族定义命令和若干条自行定义命令。  
字族定义命令为`\DeclareFontFamily{编码}{字族}{字体加载位置}`。其中，字体加载设置主要用于设定在自动断词时的连字符等内容，系统默认的连字符是`-`。  
字形定义命令的形式为`\DeclareFontShape{编码}{字族}{序列}{形状}{字体信息}{字体加载设置}`。其中，字体信息主要用于设定字体尺寸和字体文件名。  
`<->`表示所有字体尺寸，字体描述文件名后缀是.tfm，字体文件名后缀是.vf。知道字体文件名的好处是可以直接使用字体命令修改局部文本的字体。  
自定义字体命令`\font`是TeX基本命令，可以直接访问字体文件，快速高效。通常格式为：`\font\自命名命令=字体文件名at字体尺寸`。  
## 字体属性命令
`\font`字体命令时系统底层命令，不通过LaTeX的NFSS新字体选择机制，直接从字体文件中提字，在某些情况下会出现字体尺寸不一的问题。要避免该问题，应该使用根据字体的属性进行读取字体文件、显示字体文件和字体转换等字体处理操作的管理程序NFSS。
### 单项字体属性命令
LaTeX提供了5条字体属性命令和1条字体选用命令。  
1. `\fontencoding{编码}`
2. `\fontfamily{字族}`
3. `\fontseries{序列}`
4. `\fontshape{形状}`
5. `\fontsize{尺寸}{行距}`
6. `\selectfont`

可使用单项字体属性命令，但必须紧跟字体选用命令，否则无效。尺寸属性命令中，尺寸和行距两个参数的默认单位是pt，使用时可不填写长度单位。  
如果只给出字族命令而没有给出编码命令，则认为该字族的编码为当前默认编码。如果没有找到字体定义文件，系统将在编译过程文件.log中给出警告，并自动用默认的字体定义文件。这种替代可使编译不致停顿，但也会掩盖错误。  
LaTeX还提供了一条综合字体属性命令:`\usefont{编码}{字族}{序列}{形状}`。  
## 全文字体的修改
有时希望改变全文的默认字体，也就是修改常规字体。  
## 常规字体的定义
LaTeX对常规字体`\normalfont`的定义为`\usefont{\encodingdefault{\familydefault}{\seriesdefault}{\shapedefault}`。  
字族名用cm打头的都是高德纳教授在创建TeX时期开发的字族。  
## 常规字体的修改
根据系统对常规字体的定义方法，可对其进行修改。如将常规字体由罗马体改为等线体，可在导言中使用`\renewcommand{\familydefault}{\sfdefault}`。  
## 手写体
论文中很少使用手写体。但有时引用某位大师的语录、说明某一事物的自然性或随意性、或者个人签名而采用手写体，看起来会更亲切、活泼和真实。  
需要调用相应的宏包来使用手写体。  
# 中文字体
LaTeX系统不能直接处理中文，要使用汉字，必须要在导言中调用ctex宏包。  
ctex宏包提供了一个统一的中文框架，定义了一系列有关中文排版的用户命令，支持CCT、CJK、xeCJK三种中文处理方式。  
[宏包调用格式](./宏包调用格式.jpg)  
调用ctex时自动调用CJK宏包，后者是cjk宏包套件的主宏包。  
根据字符之间空格的不同处理方法，CJK宏包提供有CJK和CJK\*两种双字节文字处理环境，困在其中排版多种亚洲文字。中文默认字体是宋体，简体和繁体可在其中混合排版，因为都在同一个GBK汉字编码字体文件中。  
### 空格的处理
中文是一个汉字紧挨着一个汉字地输入，汉字之间和中文标点符号与汉字之间都不留空格，所以ctex将忽略所有汉字和中文标点符号之后的空格，即使用CJK\*环境模式，该环境删除所有CJK字符之后的空格。  
CJK\*中汉字与其他字符应有的空格也被删除了，可使用不可见空格符`~`代替应有的空格，数量比较多时，可选用ctex宏包的space选项，可保留汉字与其他字符之间的空格以及换行时自动插入的空格，即使用CJK环境。  
### 空格符和空格命令  
在LaTeX系统中，`~`是一个不可换行的空格符，即不允许在该符号所处的位置换行，如果使用过多，对LaTex的自动换行非常不利，因此ctex将其重新定义为可以换行的空格符，如果希望维持原定义，可在正文中使用`\standardtilde`来恢复。  
`\CJKtilde`将`~`定义为可换行空格符。  
如果在论文中有些中文与其他字符组成的词汇，不希望在它们之间的空格处被换行，而ctex已将`~`重新定义为可换行的空格符，这时可在空格前插入空格命令：`\nbs`或`\nobreakspace`，后者是LaTeX提供的，前者是CJK宏包提供的，是后者简化形式，二者都可生成一个不可换行的空格。  
### 两种CJK环境的转换
`\CJKspace`切换到CJK环境，`\CJKnospace`切换到CJK\*环境。  
### 中文数字
`\CTEXnumber{命令}{阿拉伯数字}`，`\CTEXdigits{命令}{阿拉伯数字}`。命令为自命名命令，且不需要预先定义，第一条命令将阿拉伯数字转换为中文数目后将其赋予命令，第二条命令将阿拉伯数字转换成中文数字后将其赋予命令。可用命令来显示中文数目或者中文数字。  
## 中文字体编码
当使用LaTeX或者PDFLaTeX编译源文件时，中文红包ctex默认的中文字体编码是GBK。使用XeLaTeX时，默认编码是UTF-8。  
## 中文字体的尺寸
CJK宏包并不附带中文字体文件，而是由CTEX配置；也不提供字体尺寸命令，通常中文字体的大小是由LaTeX的字体尺寸命令设定的，也就是说，中英文混排时，中文字体的尺寸是随同英文字体的尺寸变化的。  
## 中文字体的使用
CTeX配置了6种Windows的TrueType中文字体，分别是宋体黑体楷书仿宋幼圆和隶书，并提供了6条声明形式的中文字体命令。这些命令不可以在导演中使用，只有在CJK环境中才能被执行。  
ctex将章节命令里的中文都设置为黑体，如果希望更改某些标题的字体，就可使用中文字体命令。改变全文某个层次标题所有中文的字体，需使用ctexcap宏包。  
## 中文字号设置
国内有些出版物的字体尺寸是按字号要求的，ctex提供了一个字号命令：`\zihao{代码}`。代码取值见[字号表](./字号表.jpg)。  
英文字体的尺寸也会随同字号命令做出相应的改变，以使其与中文字体的大小保持适当比例关系。  
## 任意尺寸字体
利用字体属性命令或缩放格子命令来解决。  
如果是汉字，可以直接使用字体属性命令中的尺寸属性命令`\fontsize{尺寸}{行距}\selectfont`。通常行距要比字体尺寸大20%左右，但由于ctex已加大了行距，故保持两个参数等比即可。  
如果是英文，需要调用typelcm宏包，因为系统默认字体对字体尺寸的调正步长和上限做了规定，如果使用`\fontsize`命令。系统会在字体尺寸命令对应的字体点数数值中寻找最接近的尺寸。typelcm取消了字体尺寸步长和字体尺寸上限。  
typelcm只适用于OT1编码的字体，如果改用fix-cm宏包，还可对T1、TS1、OMS和OML编码的字符有效。为了能够确保改变系统的默认字体定义，该宏包的调用命令应为`\RequirePackage{fix-cm}`，且需置于`\documentclass`命令之前。  
ctex自动加载了fix-cm,calc,color和graphics等多个相关宏包。  
## 采用缩放盒子命令
使用宏包graphicx，并使用提供的缩放盒子命令：`\resizebox{宽度}{高度}{文本}`，如果参数为`!`，表示按照另外一个参数进行缩放同时保持高宽比不变。缩放盒子的好处是可制作不同高宽比的字体，以产生各种文字特效，而且这种方法对中文和英文都有效，无需调用其它宏包。

# 4.2 文本格式
## 断词
LaTex直接设定全文自动断词，通常无需手动设置。  
### 断词命令
若希望为某个单词设置或增加断词位置，可用`\-`以提示系统这是一个可断点。对`x-radiation`可设置为`x-ra\-di\-a\-tion`来避免换行时在`x-`之后断词。  
### 断词声明
在导言中使用`\hyphenation{单词1的断词方式 单词2的断词方式...}`来为全文中所有的相同单词设置可断点,其中`\hyphenation{Windows}`会禁用原来的可断点。  
### 禁止断词
`\uchyph=0`使其后的大写名词禁止断词。 
### 显示断词位置
`\shohyphens{单词1 单词2...}`会显示单词的可断点。  
## 连词
希望几个单词连接为一个整体，不要因为换行而被断开。可在几个单词之间加入`~`，比如Prof.~D.~E.~Knuth。  
斜杠符号也有连词的作用，比如`input/output`不会在斜杠符号处被断词，但是很有可能因此凸入右侧边空。如果希望这类词可被中断，可将斜杠符号改为`\slash`斜杠命令，但这种方法对中文无效。  
## 字距
英文单词中字母之间的距离，或是中文里汉字之间的距离。  
中文里每个汉字和符号的宽度都是相同的，因此中文字符之间的自然距离为0pt。通常无需调整英文和中文的字距就可以得到专业的排版效果。  
### 全文字距的修改
使用microtype宏包`\usepackage[参数1=选项，参数2=选项，...]{microtype}`。其对英文字母之间距离的调整功能也适用于中文。microtype宏包功能仅适用于PDFLaTex编译。  
### 局部字距修改
microtype提供字距调整命令`\textls[字距系数]{文本}`，字距系数默认是100，取值范围是-1000~1000，相当于-1em~1em。  
`\textls`不会受microtype宏包可选参数的影响，可以独立使用，并且可以用在数学模式中。该命令用在章节命令中，其对字距的调整作用也会反映到目录和页眉中的章节标题。  
### 汉字的间距
调用宏包ctex后，会自动在document环境中嵌入一个CJK环境，在该环境中，汉字之间以及汉字与中文标点之间的距离是\CJKglue，定义是`\newcommand{\CJKglue}{\hskip 0pt plus 0.08\baselineskip}`。其中`\hskip`是Tex用于水平弹性长度设置的基本命令。  
### 个别单词的字距调整
为了避免microtype的编译问题，可以使用水平空白命令来调整字间距。有`~, , \thinspace. \medspace, \thickspace, \enspace, \quad, \qquad`。
### 双引号与单引号并用
英文中双引号由两个单引号组成，在双引号和单引号并用时，应该插入适当宽度空白，否则称为三引号。  
## 连体字
系统默认字体是罗马体，此时输入的英文字母排版时大部分彼此之间有一定间隙，有少数几种字母组合排版效果是连为一体，称为连体字，主要出现在罗马体的字母拼写中，等线体和等宽体很少或没有。  
连数符合破折号也是由两个和三个连字符构成的连体字。  
连体字改变了字母的常规形态，如果不希望出现连体字，可以在其间插入一堆花括号，相当于插入一个宽度为零的字符。如`suf{}f{}ice`。  
连体字主要是某种字体的审美需要，并不是每种字体都有连体字，有些字体的连体字样式和字母组合也不尽相同。  
## 词距
指英文单词之间的距离。系统按照常规排版在单词之间插入了适当宽度的可伸缩的空白，以保证每行文本的宽度，因此通常无需调整。
1. 为增加弹性伸缩范围，可以对单词间距的长度数据命令`\spaceskip`重新赋值，该命令对中文无效。在系统中，该命令也是一个长度标志，默认值是0pt，当其为0时，使用内置的词距宽度，非零则使用作者对长度数据命令`\spaceskip`的设定值。  
2. 也可使用系统提供的字体参数命令来修改词距的伸缩范围，如将词距伸展范围扩大一倍`\fontdimen3\font=2\fontdimen3\font`, `\font`无参数独自使用，它指定的字体就是当前字体。  

### 词距补偿
由于斜体单词向右倾斜，使得右侧词距要稍微窄一些，可以使用`\/`命令，`{\it italicized\/}`，将其插在斜体单词之后，作为词距补偿。 
`\textit{文本}`本身有词距补偿，可使用`\nocorr`取消补偿。  
## 句距
指英文语句之间的距离。每当文本出现句号、感叹号或者问号并紧跟一个空格时，系统认为这是一段语句结束标志，将在词距的基础上再额外插入一段水平空白，其宽度为`\fontdimen7`。  
1. 对代表句距的长度数据命令`\xspaceskip`重新赋值可以修改句距在系统中，该命令也是一个长度标志，默认值是0pt，当其为0时，使用内置的句距宽度，非零则使用作者对长度数据命令`\xspaceskip`的设定值。  
2. 也可使用`\fontdimen7`修改额外插入的空白宽度，如将句距宽度扩大一倍`\fontdimen7\font=2\fontdimen7\font`, `\font`无参数独自使用，它指定的字体就是当前字体。  

## 句号后的空白
1. 紧跟在大写字母后的句号，系统不视其为句尾，仅在其后插入词距所需空白。
2. 缩略语或计量单位等大写字母就在句尾时，在大写字母与句号之间插入句尾命令`\@`或`\null`。
3. 并非所有小写字母之后跟句号就是句尾，在句号后加入空格命令`\ `告诉系统这里按词距处理。  
除参考文献外的非英语文本排版中，经常要取消所有句间附加空白，即使一行中的所有词距相等，可在其前加入`\frenchspacing`，这样`\@`命令也就不需要了，可用`\nonfrenchspacing`恢复默认设置。  

## 换行
经过编译得到的PDF文件，其行宽是导言中版面尺寸命令或版面设置宏包设定的。系统会根据设定的版心宽度，自动进行换行，必要时还可按照设定的断词规则，在换行处自动进行断词处理，通常无须干预系统的自动换行。  
### 立即换行
换行命令`\\, \\*, \\[高度], \newline`。第二个为立即换行，但不能在此换页。第三个为换行且当前行与下一行之间增加一段高度的垂直空白，多用于调节表格数据行之间的间隔，或多行公式之间的间隔。第四个只能用于段落模式，其它三种还可用在某些数学环境中。  

### 智能换行
`\linebreak[优先级]`让系统根据情况确定是否提前换行，且换行后不给当前行留下空白段。优先级分为0~4,0是作者既不反对也不鼓励，默认是4级。该命名在换行时会将当前行中的英文词距和中文字距拉宽，以弥补当前行剩余的空间。  
### 禁止换行
`\nolinebreak[优先级]`，优先级分为0~4,。实际中，优先级为0~3时，通常都被忽略。  
`\linebreak和\nolinebreak`只能在段落模式中使用。  

# 空格与空行
空格可以由空格键或Tab键产生。  
1. 每行文本之间的空格都将被忽略
2. 两个单词之间的所有空格，只保留一个，其他的都被忽略。
3. 文本中保留的所有空格都转换为弹性空白，以使每行文本两端对齐。  
4. 按一次回车键仅相当于插入一个空格，尽管源文件再次换行，但系统不会再次换行。  
5. 连续多个空行，系统只视其为一个空行，即只相当于一个换段命令。

# 标题
标题是表明某一论述内容的简短语句，是对论述的高度概括。  
论文写作中要用到三种类型标题：论文题名(title)、层次标题(heading)和图表标题(caption)。  
## 论文题名
提名应内容简短准确地概括论文的核心内容和主要论点，题名所用的词语应便于作为文献数据进行著录和检索。长篇论文有单独题名页，通常是论文的封面；短篇论文的题名一般与正文相连。  
两种创建论文提名的方法：  
1. 一组题名信息命令生成论文题名，即可单独一页，也可和正文相连。
2. 用题名页环境声场单独的题名页。  

很多论文要求双栏排版，但是论文题名总是以通栏即单栏的格式排版。  
## 题名信息命令
`\title, \author, \date`等，`\maketitle`是题名生成命令，根据其它命令中的信息，自动进行排版，必须位于其它题名信息命令之后，没有该命令，其它题名信息命令都无法生效。  
`\maketitle`由所选文类提供，不同文类对信息题名信息的处理不同，book和report默认创建单独的题名页，使用`notitlepage`选项更改，article默认题名与正文相连，使用`titlepage`选项更改。  
在独立的题名页中，用致谢命令生成的脚注不附带脚注线，脚注序号是阿拉伯数字，脚注字体尺寸为`\small`，在题名与正文同页面中，致谢命令生成的脚注带脚注线，脚注的序号是星号等脚注标识符，字体尺寸为`\footnotesiz`。  
使用题名生成命令(`\maketitle`)生成题名信息，必须使用题名命令(`title`)，作者姓名也要使用`\author`编写，其它题名信息可以选用。修改日期格式可以使用datetime宏包。  
题名信息命令好处是方便，但是不易对题名页样式做大范围的改动，如果希望对论文封面进行艺术设计，最好采用titlepage题名页环境。  
## 题名页环境titlepage  
创建的提名页面没有页眉和页码，其后页面的页码为1，`\today`可用于该环境中，其它题名信息命令在环境中无效。  
好处是灵活性强，但只能生成独立的题名页，即使用`notitlepage`选项也不予理睬。  
## 层次标题
系统中各层次标题都是由所选文类提供的各种章节命令生成的。book和report两种文类有以下7种章节命令，用于生成7个层次的标题：  
1. `\part{标题}`
2. `\chapter{标题}`
3. `\section{标题}`
4. `\subsection{标题}`
5. `\subsubsection{标题}`
6. `\paragraph{标题}`
7. `\subparagraph{标题}`  

article有6种，缺少章命令`\chapter`。  
层次标题基本格式：  
标题标志 标题内容
标题标志：可包括层次名和序号两个部分。例如Chapter是层次名，1、1.1等是序号。章标题一下各层次标题的序号都是由上一层标题的序号，加一圆点，再加上该标题在本层的序数组合而成。  
book和report文类对各层标题的定义如下：  
* 部和章标题具有完整的标题标志，即有层次名和序号。  
* 节标题和小节标题有序号无层次名。  
* 小小节及一下标题无标题标志，及无层次名和序号。  
* 小节及以上层次的标题可以被排进章节目录。  
* 章标题和节标题还可被写入页眉。  

### 章节命令的功能
章节命令有以下功能：  
1. 不同层次的章节命令，可自动设置相应的标题格式，如字体尺寸等，同层次的标题格式相同。  
2. 同层次的标题与上下文间的空白宽度相同。当节标题或节以下标题出现在页面顶部时，该标题与上文的垂直空白被取消。  
3. 所有层次标题不会被排版到页面的底部，即版心的最后一行。  
4. 每种章节命令都有与其命令名相同的序号计数器，每执行一次章节命令，其序号计数器加一，并将低于其一个层次的序号技术器清零。  
5. 自动为符合设定排序深度的层次标题编排序号，这些层次标题的增删或前后位置调整不会影响排序的正确性。  
6. 各种章节命令之前或之后的空行都将被忽略。  
7. 章命令和节命令可将标题内容写入页眉；采用article文类时，节命令和小节命令可将标题内容写入页眉。  
8. 将层次名、标题序号、标题内容和标题所在页的页码写入.aux引用记录文件，如果是子源文件，在分段编译时这些标题信息数据将被调入.toc章节标题记录文件。  

### 章节命令的使用
book和report提供的`\part`可生成单独的部标题页，article提供的`\part`则不专设一页。  
部标题作用是将有很多章的长篇论著分成若干部分，排序采用罗马数字。不参与其它层次标题的排序，其它层次标题使用阿拉伯数字排序。  
除部命令外，其它所有章节命令都应从高到低逐层使用，中间不得空缺。  
book和report提供的`\chapter`都将新开一页，book默认为每一新章都从奇数页（右页）开始，可启用openany选项避免。report默认新章可从任意页开始，如需要统一从右页开始，可使用twoside和openright选项。  
book通常写作学位论文，要用到章标题；学术论文大都对篇幅有所限制，一般最高到节标题，常用article文类。  
### 章节命令的可选参数
每种章节命令都有一个可选参数，例如：`\section[目录标题内容]{标题内容}`。  
通常省略可选参数，如果给出，则标题内容只排版到论文正文中，目录标题内容排入目录和页眉中。有时标题很长，使用可选参数来简化标题内容。 如果标题内容含有参考文献引用命令或脚注命令，章节目录或页眉中也会有参考文献的引用标号或脚注序号，可使用可选参数解决。  
### 带星号的章节命令
带星号的形式生成的层次标题没有标题标志，也不被排进目录和页眉。目录和参考文献的标题就是系统使用`\chapter*`或`\section*`生成的。  
若希望带星号的章节命令生成的摘要、参考文献或索引也能被排进目录。可在章节明令或环境命令后紧跟一条添加条目命令`\addcontentsline{toc}{层次名}{标题内容}`。若需要排进页眉，紧跟一条双标题命令`\markboth{章标题内容}{节标题内容}`。两个参数只用其一，另一个刻意空置，如果只是节标题，也可用`\markright{节标题内容}`。  
所有带星号的命令没有可选参数，也不会触动同名的序号计数器。  
### 章节命令中添加脚注命令
使用footmisc宏包，`\usepackage[stable]{footmisc}`，否则编译错误。  
### 反斜杠
标题内容有反斜杠时，不能使用抄录命令来限制其作用。可用`\string`字符串命令。但这种方法无法正确传递给章节目录，可将带反斜杠的字符串分解为两组字符串来解决，`\symbol{当前字体中反斜杠的十进制代码}`。这种方法对书签仍然行不通，可使用hyperref提供的双字符命令`\texorpdfstring{LaTex字符}{PDF字符}`。其中，LaTex字符是用于章节命令中的字符和各种命令，PDF字符是可在书签中使用的字符和命令。  
### 章节命令与层次深度
[章节命令与层次深度](./章节命令与层次深度.jpg)   
`\setcounter{secnumdepth}{深度}`。  
### 标题的引用
使用引用命令`\ref`得到的只是章节或图表标题的序号。使用`\titleref{书签}`，就可以引用章节标题或图表标题的标题内容，如果还需添加标题序号，可对`\theTitleaReference`重新定义。  
## 层次标题格式的修改
章标题的标题标志与标题内容分为两行，章标题与下文的距离较大，所有层次标题都是左对齐，英文字体是直立形罗马体。LaTex并没有提供任何专用于修改标题个格式的命令。  
通常做法是：如果调用了ctex宏包，就改为调用中文标题宏包ctexcap来修改标题格式。纯英文使用titlesec标题宏包。  
titlesec提供了一条标题格式命令`\titleformat{章节命令}[形状]{标题格式}{标题标志}{间距}{标题内容}[后命令]`。该命令虽然强大，但是过于复杂，因此宏包提供了一个简化的标题格式命令`\titleformat*{章节命令}{标题格式}`。  
### 标题周矩设置
`\titlespacing*{章节命令}{左间距}{上间距}{下间距}[右间距]`。通常上下间距都设置为弹性长度，但是命令的输入和阅读都不方便。可用`\titlespacing{章节命令}{左间距}{*n}{*m}[右间距}`。上下间距分为别nex plus 0.3ex minus 0.06ex和mex plus 0.1ex。  
### 标题线的设置
`titlerule[高度]`和`\titlerule*[宽度]{字符}`。前者可用在标题格式命令或标周轴距命令中画水平实线，后者可用在标题格式命令或标周轴距命令中画水平虚线。  
### 带星号章节命令的修改
上述对章节命令的修改也等效于对带星号的章节命令的修改，所以会影响到目录、参考文献等，因为这些无标题标志的标题是由系统在其内部使用`\chapter*`命令生成的。如希望保持原来格式，可将对章命令的修改命令放到正文中，且在章节目录命令之后。  
### 预定名
自动生成的层次名是系统使用预先定义的的预定名生成的。  
## 中文标题宏包ctexcap
调用ctexcap宏包时，自动加载ctex中文字体宏包，如果此前已经调用了ctex，应将其转移到ctexcap的调用命令中，而将原有的ctex命令删除，以免相互冲突。  
### 预定名的修改
`\CTEXoptions[参数1={中文名},参数2={中文名}]`  
### 层次标题格式的设置
在调用ctexcap宏包后，它还会自动将各种层次标题按照中文的阅读和排版习惯重新设置，使用nocap选项，将恢复所有预定名和层次标题的原貌。如果希望修改层次标题的中文默认格式或nocap选项格式，可使用标题格式命令`\CTEXsetup[参数1={格式}，参数2={格式},...]{层次名}`。  
使用`\CTEXsetup`时，将完全覆盖该参数的原有设置。如果只是在原有设置基础上，对部分设置进行修改，可以使用带“+”号的参数。如：`\CTEXsetup[format+={\fongsong}]{section}`。  
### 附录的预定名修改
文类book和report中的附录预定名是借用章命令生成的，其序号是用chapter计数器编排的，而article文类采用的是节命令和section计数器，且无预定名，只有序号。附录相当于book、report中的一章，article中的一节；修改章或节标题的格式也就修改了附录的标题格式。  
# 5.4 图表标题
一段文字到盘面底部没有排完，系统可将剩余部分移至下一页的顶部继续排，而一个插图或表格被整体装入一个盒子不能拆分，如遇当前页剩余空间排不下时，系统只能将整个插图或表格移至下一页的顶部，这会给当前页留下大片空白，而且图表标题也被打乱。  
## 浮动环境
使用figure图形浮动环境，table表格浮动环境和`\caption`图表标题命令解决上述问题。  
`\begin{figure}[位置] 插图命令或绘图环境  \caption[目录标题内容]{标题内容}\end{figure}`  
`\begin{table}[位置]\caption[目录标题内容]{标题内容} 表格环境 \end{table}`  
这两个环境可以根据其中图标的高度结合版面编排情况，确定图表在版面中的位置，而图表前后的文本，也可推后或前移，以填补图表“浮动”产生的空间。图表标题命令`\caption`所生成的图表标题可随同图表一起浮动。  
浮动环境将其中所有插图或表格放在一个宽为`\linewidth`的`\parbox`段落盒子中，构成一个不可分割的，但可在版心中沉浮的盒子，所以在浮动环境中的插图或表格称为浮动体。  
浮动环境中的内容将按照导言中的格式设置进行排版，在正文中的各种格式设置对齐无效。如果要改变浮动环境的排版格式1，必须将相应的设置命令置于浮动环境内部。  
解说图表时，不宜使用“下图”、“上表”之类的定位词，而应使用图表的标题标志。最好使用交叉引用生成其中的序号。  
### 浮动体的放置原则
浮动体的位置参数是作者对浮动体摆放位置的建议，但系统有否决权。这有不违背一下原则时，系统才采纳作者的建议。  
1. 浮动体不能放置在其浮动环境所在版面之前，即浮动体可以向上、向下和向后浮动，但不能向前浮动。  
2. 插图浮动体按出现的顺序放置，不能先后颠倒，表格浮动体也是如此，但这两种浮动体之间允许错位。在双栏版面中的两种跨栏浮动体也允许错位放置。  
3. 如果所给位置参数的选项是ht，那么系统优先考虑h的可能性，即就是在版面顶部有足够的放置空间。  
4. 不能因放置浮动体而造成版面溢出，即超出版心的侧边或底边。  
5. 遵守为保证排版美观而对浮动体放置所采取的限制措施，但如果给出!位置选项，将部分取消对文本页中浮动体的限制。

### 双栏版面的跨栏浮动环境
尚书浮动体环境适合于浮动体宽度小于或等于版面宽度或一栏的宽度。若双栏排版照片那个，浮动体的宽超过栏宽，或希望浮动体置于两栏之间，要使用带星号的浮动环境。  
`\begin{figure*}[位置] 插图命令或绘图环境  \caption[目录标题内容]{标题内容}\end{figure*}`  
`\begin{table*}[位置]\caption[目录标题内容]{标题内容} 表格环境 \end{table*}`  
其可选参数位置只有t和p两个选项。  
## 图表标题命令
`\caption[目录标题内容]{标题内容}`，其中，目录标题内容只出现在插图目录或表格目录，默认值是标题内容。
生成的插图或表格标题的格式为：
标题标志 分隔符 标题内容，各部分说明如下：
> 标题标志 由标题名和序号两部分组成。figure环境中，标题名是Figure,序号是由figure计数器编排；table环境中，标题名是Table，序号由table计数器编排。采用book/report文类，序号由章序号加一圆点再加该图或表格在本章的序号组成；如果是article文类，序号以全文为排序单位 
分隔符 默认值为一冒号和一空格

图表标题不足一行，居中排版，否则按照首行不缩进排版。
### 标题名的修改
Figure和Table也是系统预先定义的预定名，定义命令分别是:`\figurename`和`\tablename`，可重定义来修改标题名。  
### 标题位置的调整
图表标题与其上面的图表或上文之间距离为`\abovecaptionskip`,默认值是10pt,与其下的距离为`\belowcaptionskip`，默认值为0pt。 
# 图表标题格式的修改
## 排序单位的修改
可使用amsmath宏包日功德排序单位命令，如`\numberwithin{figure}{section}`。若已调用caption宏包，也可使用其序号选项来改变。  
## 图表标题宏包caption
该红包允许在浮动环境之外使用`\caption`。  
`usepackage[参数1=选项，参数2=选项，...]{caption}`，如需对正文中某些图表标题格式进行设置，可用caption提供的标题设置命令`\captionsetup[浮动体类型]{参数1=选项，参数2=选项，...}`,浮动体类型默认为图和表。  
## 标题设置命令
`\caption*{标题内容}`:可生成无标题标志的图表标题，标题内容也不会被编入插图目录或表格目录。    
`\captionof{浮动体类型}[目录标题内容]{标题内容}`:可为浮动体环境之外的图表生成标题。  
`\captionof*{浮动体类型}[目录标题内容]{标题内容}`:同上，但无标题标志。  
`ContinuedFloat`:使一组相关插图或表格的标题使用同一序号。  
`ContinuedFloat*`：可在序号后加一子序号，会与subfig子浮动体宏包冲突。  
ContinuedFloat是caption宏包定义的计数器，默认值为空。  
## 图表的侧标题
caption宏包可对图表标题格式进行全面设置，但无法改变标题与图表的相对位置。解决要使用floatrow宏包。  
`usepackage[参数1=选项，参数2=选项，...]{floatrow}`  
`\floatsetup[浮动体类型]{参数1=选项，参数2=选项，...}`   
图标默认对齐方式是左对齐，使用floatrow宏包后，全文所有浮动环境中图表都居中对齐，除非另有对齐命令；所有浮动环境中的标题都置于图表之下，但是表格标题通常位于表格之上，可使用`\floatsetup[table]{style=plaintop}`解决。  
### 侧标题设置
使用侧标题命令`\fcapside[宽度]{标题命令}{插图或表格}`，实际是将浮动环境排版为两栏：标题栏或表格栏，两栏之间的空白宽度可以用参数capbesidesep设定。  
宽度参数用于设定图表栏的宽度，如为`\FBwidth`表示测标题栏的宽度=行宽度-图表宽度-栏间空白宽度。如果在导言中调用了calc算数宏包，宽度参数可使用长度表达式。  
默认情况下`\fcapside`将浮动环境排版为等宽的两栏，标题和图表分别水平居中。  
如果侧标题的位置和奇偶页有关，那么源文件需经过连续两次编译才能得到正确结果：第一次确定侧标题所在页码，并写入同名的.aux文件。第二次根据页码的奇偶，确定侧标题的排版位置。  

# 8.1 数学模式
数学模式分为行内公式（inline math）和行间公式（display math）。前者将数学式插入文本行之内，使之与文本融为一体，适合编写简短的数学式。行间公式将数学式插在文本行之间，自成一行或一段，与上下文附加一段垂直空白，使数学式突出醒目。多行公式、公式组和微积分方程等复杂的数学式都是采用行间公式形式编写的。  
## 8.1.1行内公式
LaTex提供了三种方法编写行内公式：
1. `$...$`;
2. `\(...\)`;
3. `\begin{math}...\end{math} `

既可用于段落模式也可用于左右模式。段落中的行内公式应简短，以避免中断换行，因为系统只能在某些数学符号处自动换行。  
若禁止在关系符或二元符处换行，可使用`\relpenalty=10000`或`\binoppenalty=10000`赋值命令。这两个控制命令默认值分别为500和700,取值范围为-10000~10000,代表从必须到禁止的坚决程度。也可在禁止换行处插入`\nobreak`命令。  
可以使用`{}`包含数学式，避免换行  
换行位置附近无二元符，行内公式无法中断，此时可以使用分段处理的方法，即将一个`$$`替换为两个`$$`。  
三种行间公式:  
1. `$$...$$`;
2. `\[...\]`;
3. `\begin{displaymath}...\end{displaymath}`

第三种最常用，第一种是Tex的原始方法，在有些特殊情况下会出问题，比如不能执行标准文类fleqn公式左缩进对齐选项等，但也有个别情况只能使用第一种方法。  
在数学模式中不得有空格或\par换段命令，否则系统提示出错。  
`equation`和`eqnarray`分别用于编写带有序号的单行和多行行间公式。  
# 8.2 常用数学宏包
## 宏包套件ams
调用公式宏包amsmath时，amsbsy, amsgen, amssopn, amstext四个宏包也会自动加载。  
ams提供有amsart, amsproc, amsbook三个文类，用于期刊、学报和专著的排版。   
## amsmath选项
intlimits指定行间公式中，积分符号的上下标分别置于其顶部和底部。  
nointlimits默认值，无论行内公式还是行间公式，积分符号的上下标都置于右侧。  
namelimits默认值，行间公式中，det, inf, lim, max, min等算符的下标置于底部。  
nonamelimits，无论行内行间，上述算符的下标都位于右侧。  
sumlimits行间公式中，将大型符号的上标、下标分别置于顶部和底部。  
nosumlimits无论行内行间，上述符号（不包括积分号）上下标都置于右侧。  
## 宏包套件amsfonts
amsfonts是AMS根据其印刷和电子出版物以及在线资料库的字符样式要求，编造的一组用于排版数学出版物的数学符号和字体宏包，这些数学字符都是LaTex所没有的。  
amsfonts宏包套件包括amsfonts, amssymb, eufrak等宏包。其中调用amssymb时同时加载amsfonts。  
公式宏包amsmath和字符宏包amssymb是编排数学式时最常用的宏包。  
## 宏包套件mh  
mh由以mathtools宏包为主的多个数学宏包组成，在amsmath的基础上又提供了许多数学符号和排版命令。  
调用mathtools时，自动加载amsmath, calc和graphicx等宏包。  
## 一些命令
`\text`可以把简短文字插在行间公式任意位置。 
`\boldmath`粗体数学字体命令，为声明形式，必须用在数学模式之外，如`{\boldmath $\Gamma+\Delta$}`。  
`\quad, \qquad, \!, \,, \:, \;`是水平空白命令，即可在数学模式中使用，也可在文本模式中使用。  
## 省略符号  
* `\cdots`水平位置与减号等高。  
* `\ddots`对角省略号，多用于矩阵环境。  
* `\ldots`可用于文本模式，水平位置与基线平齐。  
* `\dots`可用于文本模式，功能与`\ldots`完全相同；但调用amsmath后，该命令在数学模式中，可根据前后的算符自动地确认省略号的垂直位置。  
* `\vdots` 垂直省略号。  
* `\hdotsfor{n}`须调用amsmath，用于矩阵环境，可横跨n列的省略号。  
* `\iddots`反对角省略号，须调用mathdots宏包。  

# 8.3数学符号
可以在数学模式中使用的符号生成命令简称数学符号。 
LaTex中大写希腊字母是直立的，小写希腊字母是斜体的。  
函数使用其符号命令生成，可得到直立罗马体函数名和适当的数学间距。
### 斜体大写希腊字母
可在数学模式中使用`\mathnormal`或`\mathit`，也可以使用amsmath提供的符号命令，就是在希腊字母命令前加"var"，如`$\varGamma$`。  
### 粗斜体希腊字母
使用amsmath提供的粗体符号命令，`\boldsymbol`。
### 直立小写希腊字母
使用宏包upgreek,在小写的希腊字母生成命令前面加一个"up"，如`$\upmu$`。  
如果要消去数学式的某一项（斜向右上的线段），可用cancel宏包中的`$\cancel{(a+b)}$`。
### 不可断词的连字符
为了防止在有连字符的数理名字的连字符处断词，amsmath提供了一条禁止断词命令`\nobreakdash`。  
### 新算符定义  
`\DeclareMathOperator{算符命令}{算符}`，带星号的形式，可在行间公式中将新算符的上下标从右侧移到顶部和底部。  
使用新算符有3个好处：  
1. 便于编排数学式  
2. 可与单个斜体字母变量明显区别开  
3. 系统会自动在新算符的前后添加适量的数学空白  

### 四角标定义
amsmath提供了一个四角标定义命令:`\sideset{^左侧上角标_左侧下角标}{^右侧上角标_右侧下角标}`，在四角标定义命令之后只能使用`\sum, \int`等大型运算符号。  
### 上下标
LaTex系统提供的`\stackrel{上标}{字符}`以及amsmath提供的`\overset{上标}`, `\underset{下标}`也都有将上标或下标放置在指定字符的顶部或底部的功能。   
### 边框
在$\int^{b}_{a}f(x)\,dx=-\int^{a}_{b}f(x)\,dx$中，被积函数与积分算子dx之间应用`\,`插入一小段空白。  
可以使用`\fbox`,想要修改公式两侧与边框之间的空白时，可以在边框盒子命令前添加`\mathsurround=xpt`，若将边框盒子命令插入其他公式环境，可生成带有边框的行间公式，如
```latex
\begin{equation}
\fbox{$\int^{b}_{a}f(x)\,dx=-\int^{a}_{b}f(x)\,dx$}
\end{equation}
```
也可使用amsmath提供的`\boxed`,使边框成为数学模式，当插入其他公式环境，可成为带边框的行间公式，同时字体尺寸也随之增大。  
`\Aboxed`命令可在多行公式中为某一行公式生成边框。  
以上边框都是左右盒子，只能放置但行文本或数学式，且不能设定边框的宽度。如果需要设定统一的边框宽度，或者放置多行文本或数学式，需要调用everb宏包，并使用其提供的colorboxed环境，即`\begin{coloredbox}[边框]数学式\end{colorboxed}`，其中边框可选参数有很多可选子参数，不仅可用于设定边框的宽度，还可以设置边框各部位的颜色。调用everb宏包时，xcolor宏包也被自动加载。  
colorboxed生成的边框可以跨页。  

# 8.4公式环境
LaTex提供有equation单行公式环境和eqnarray公式组环境，前者已被amsmath公式宏包重新定义，后者可以用排版效果更好的align, multiline或equation+split环境替代。  
LaTeX提供了一个数组环境array，数组宏包array对该环境做了功能扩展，使其具有更多的列格式选项。  

|环境名|用途|
|-|-|
|align|公式组环境|
|alignat|公式组环境|
|flalign|同上|
|gather|同上|
|aligned|块环境|
|alignedat|同上|
|cases|左花括号环境，无序号|
|equation|单行公式环境|
|gathered|块环境，无序号|
|multline|多行公式环境，无序号|
|split|多行公式环境，无序号|
|subequation|子公式环境|

除子公式环境外，所有排序公式环境的序号都由equation计数器提供。  
## equation环境
可将一个公式，无论多长，排版为一行，并给出一个序号。系统提供的displaymath环境等效于宏包提供的equation\*环境。二者都只能用于编排单行公式，其间，换行命令无效，换段命令非法。  
## array环境
数组宏包array对系统提供的数组环境array做了扩展:`\begin{array}[位置]{列格式} \end{array}`。  
格式参数说明：  
1. \*{n}{列格式}n个列格式相同的相邻列  
2. |一条垂直线 ||两条并排的垂直线，间距可用双线距离命令`\doublerelesep`来设置  
3. @{声明}也称为@表达式，将表格侧边与列之间的每行空白，或列与列之间的每行空白删除，插入声明内容，例如字符等；如果仍需要一定宽度的空白，可在声明中加入`\hspace{宽度}`，如果声明为空，那么只删除空白。可在声明中使用附加宽度命令`\extracolsep{宽度}`，它将其后与下一个`\extracolsep{宽度}`命令之前的所有列间插入长为宽度的水平空白；可以在第一列的列格式中插入`@{\extracolsep{\fill}}`使得所有列在设定的表格宽度中均匀展开。  
4. \>{声明}用在l, c, r, p, m, b之前，将声明内容插在该列所有数据之前，声明中还可以使用`\kaishu`等命令，改变该列所有数据的字体。  
5. <{声明}同上，插入位置为该列所有数据之后。  
6. !{声明}可在列格式的任何位置使用，将声明内容插入所在位置。  

列格式有左、中、右三种对齐方式，位置和列格式的各种选项及功能和表格环境tabular完全相同，array就是数学模式中的表格环境。  
非常适合多行公式的选择性对齐，还可以在列格式中使用@表达式。  
可用表格线划分数组，还可以在数组环境之外使用各种括号。  
可将array环境插入equation单行公式环境，使其能编排多行公式。  
array数组环境中，列与列分割线之间的水平距离是`\arraycolsep`，默认值是5pt，两列之间的空间就是10pt，有时此宽度显得过大，可使用长度赋值命令来修改。  
LaTeX不能对较长的行间公式自动分段，虽然breqn宏包有自动分段功能，但其分段处未必是理想位置，最好还是由作者根据公式结构，通常在+、-号处手工分段。  
`-`旁边加上{}可以保留其作为减号的两侧空白，而不是按照负号处理。  
### 取消数学模式
在array环境中使用`>{$}c<{$}`列格式，可以生成一个文本模式的列，因为增加的$抵消了内在的数学模式符号$，如：  
```latex
\begin{equation}
\begin{array}{c|>{$}c<{$}}
a_{11} & 数组A \\ \hline
a_{21} & 数组B
\end{array}
\end{equation}  
```

### 生成符号列  
列格式使用@表达式，生成一个符号列，如：  
```
\[y = \left\{ \begin{array}  
{r@{\quad:\quad}l}
-1 & x < 0 \\
0 & x = 0 \\
1 & x > 0
\end{array} \right. \]
```
## 公式组环境eqnarray
由系统提供，默认对齐方式为右对齐。  
如果需要以等号对齐，需要在等号两侧加入&列分隔符号。插入&会增添附加空白，可用赋值命令如`\arraycolsep=0.15em`来消减。  
使用`\lefteqn`可以将某一行公式左对齐，并视其宽度为0。  
宏包subeqnarray提供的subeqnarray环境与eqnarray环境的功能相似，只是生成的公式序号改为3.6a这种形式。  
宏包eqnarray将可排序的eqnarray环境与可设置列格式的array环境结合，定义了equationarray环境。  
## 公式组环境align
用于将公式组或者多行公式关于某个字符对齐。  
它以&为分列标志，奇数列右对齐，偶数列左对齐，奇偶列并肩对齐，形成“列对”。列对之间的空白宽度与列对两侧的空白宽度相等。  
## 公式组环境flalign  
与align功能基本相同，唯一区别是列对之间的距离为弹性宽度，以使公式组两端对齐。  
align,flalign环境中，最短列对间距为`\minalignsep`，默认值为10pt，可用重新定义命令来调整列对之间的最短距离。  
## 公式组环境alignat  
与align功能相似，不同在于列对之间的默认距离为0pt，通常都要插入水平空白命令，以控制列对之间的距离；另外还多了一个参数项，用于设置列对的个数，其数目可大于实际值，但不能小于。  
## 公式组环境gather
用来编写中心对称的公式组，以换行命令来区分各个公式，每个公式都与公式行居中对齐，都有自己的序号。  
## 子公式环境subequations
子公式环境将其中所有的公式视为一个整体，给出一个主序号，而其中每个公式的序号为主序号+自序号。  
子公式环境中可以嵌套align, equation。  
子公式环境可跨越任意文本和各种公式环境。  
引用子公式环境命令后的书签，得到的是主序号；引用子公式后的书签，得到的是子公式的序号。主序号是由公式宏包amsmath定义的parentequation计数器给出；子序号是由系统的equation计数器提供。  
## 多行公式环境multline  
可用于编写多行公式，首行左对齐，尾行右对齐，中间行居中对齐，整个公式的序号在尾行给出。  
可以用`\shoverleft`或`\shoveright`使中间行左对齐或右对齐。  
长公式如何分行没有一定之规，尽管行内公式都是在关系符或二元符之后换行，但行间公式通常是在它们之前分行。  
## 多行公式环境split
用于排版多行公式，与multiline区别主要为：  
1. 用&作为分隔符，但至多两列；左列右对齐，右列左对齐，形成一个列对，可使多行公式关于某个符号垂直对齐。因为它排版的多行公式更为整齐美观。如果不使用&，那么所有公式为一列，且全都与首行公式的右端对齐。  
2. 必须置于除multiline环境之外的其他公式环境中。  
3. 自身并不生成公式序号，而是由外在公式环境提供，序号垂直居中。  

## 块环境gathered, aligned, alignedat
gather, align, alignedat都可以排版多行公式或公式组，但每行公式都占据了公式行的总宽度，即当前文本行的宽度，也就是说无法将它们作为一个“块”，放入其他公式环境中，成为其他公式的一个组成部分。  
amsmath在上述三种环境功能的基础上又分别派生出三种名字带"ed"的环境，特点是每行公式只占据自身的自然宽度，这样可在一行中放置多个公式块。所有块环境都不提供序号。  
左、右命令`\left`和`\right`必须成对使用，如只需一个，另一个需要空置。

# 9插图
LaTeX的绘图功能比较简单，虽然可以使用众多宏包，但是缺点是作图不直观，绘图命令繁多，不易熟练掌握。现在通常是先把图形画好，之后再插入到LaTeX源文件中。  
## 图形的种类
两大类：位图图形和向量图形。  
位图分为有损压缩格式和无损压缩格式，TIFF、PNG和GIF是无损压缩，JPG是有损压缩。  
向量图形都是用数学式表示和存储的，其尺寸一般都较小。因为计算机显示器只能用网格显示图像，所以无论是向量图形还是位图图形，在屏幕上均以像素的方式显示图形。  
## 位图与向量图的比较
位图作为插图时要以实际大小插入源文件，不要用插图命令将大图缩小，这样既浪费内存，又会造成图形失真。向量图形的显示速度比位图慢得多，因为有一个向量计算的过程。  
内容复杂的图形或者照片适合用位图，图表和工程设计图适合用向量图形。  
## 图形格式
可直接在源文件中使用的图形格式有以下几种：
1. EPS（Encapsulated Post Script）,内嵌式Postscript语言文件格式，是一种混合图形文件格式。如需打印EPS文件，必须使用Postscript打印机，几乎所有绘图和排版软件都支持EPS格式的图形。  
2. PS,专门为打印图形和文字而设计，可以精确地描述任何平面文字和图形。其输出不依赖输出设备或输出介质，且独立于OS。PS文件以文本方式存储，与HTML相似，可用写字板打开编辑浏览。PS将文本和图形用直线和三次贝塞尔曲线表示，可任意地移动、缩放或旋转。  
3. JPG(JPEG)，有损压缩格式。采用平衡像素之间亮度色彩的压缩算法，有利于表现带有渐变色彩且没有清晰轮廓的图像。JPG图形文件在打开时自动解压缩。  
3. PNG(Portable Network Graphics)。PNG保留灰度和RGB图像中的透明度，可使图形与背景融为一体，显示速度很快，只需下载1/64的图像信息就可以显示出低分辨率的预览图像，还支持透明图形操作。  
4. PDF，基于PostScript成像。PDF相比PS，优势在于可对图像和文本进行压缩。

## 图形的插入
需要调用基于graphics的graphicx宏包，该宏包具有多个选项，常用的有`draft`和`final`，前者将正文中所有插图用与其外形尺寸相同的方框代替，在方框内只显示插图名，这样可以加快显示或打印速度，大幅度减小文件的占用空间。后者是默认值，用于抑制文档类型命令中通用选项draft对插图宏包的作用。  
## 插图命令
`\includegraphics[参数1=选项，参数2=选项,...]{插图}`，可选参数分为外形参数、裁切参数和布尔参数三类子参数。  
`origin`参数可以设定插图旋转点的位置，默认值是插图的基准点。  
[图形中旋转点位置示意图][1]  
如果在插图命令中只设定了插图的高度或者宽度，那插图的宽度或者高度将按原图比例进行缩放。  
如果没有使用外形参数来设定插图的外形尺寸，那么插图命令将按图形的实际尺寸把图形插入到文件中。  
由于插图命令既不会结束当前行也不会结束当前段，所以可把图形插入某一段落的文本行内。但通常无论在段落之中或在段落之间，通常都将插图置于单独一行之内。  
论文中的每幅图都应在其前的正文中明确提及，即每幅插图的序号都应在其前的正文中被引用。因此，插图都是置于图形浮动环境，这样可使用图标标题命令为插图生成带有序号的标题，也可在该命令中使用交叉引用命令，还可将插图标题编入插图目录。  
## 选项顺序与排版效果
`\includegraphics`从左到右处理命令中各个参数，因此选项的先后顺序可能导致不同的排版效果。  
## 带星号的插图命令
若要不显示用viewport或trim参数设定的显示区以外的图形，也可不使用clip参数而用带星号的插图命令。`\includegraphics*`。  
## 旋转点的选择
1. 表示旋转点的位置的两个字母可以任意排序。
2. 字母c是表示水平中线还是垂直中线，要视与其组合的字母而定。
3. 只指定一个字母，则另一个默认为c。  

## 插图尺寸的运算
如调用calc宏包，还可在插图命令中使用算术式来设置插图的外形尺寸。  
## 参数的统一设置
若论文中插图很多，切某些参数的设置完全相同，可用`\setkeys`统一设置参数。  
`\setkeys{Gin}{width=0.8\textwidth,height=35mm}`，其中Gin是graphicx宏包定义的标识符。该命令由graphicx在加载附带的keyval宏包提供。keyval是宏包套件graphicx的成员之一。主要用于执行`\includegraphics`命令的各种子参数功能，也可用于对其他宏包命令中的子参数执行。  
## 插图搜索
插图多的时候，通常按章存放。插图较少，整篇论文可用一个文件夹存放插图。此外有些插图是从其他文件夹中借用的。graphicx宏包提供了一条插图路径命令:`\graphicspath{{路径1}{路径2}{路径3}}`。  
每个路径都用花括号括起来，即使只有一个路径也不例外。默认搜索路径是源文件所在文件夹。`{directory name/} = {./directory name/}`。路径分隔符要使用`/`。  
如果插图是整页的外部PDF格式文件，应该使用pdfpages宏包提供的插图命令。`\includepdf[pages={3, 5}]{book.pdf}`，还可以使用合并命令`\includepdfmerge`将多页外部文件自动缩小合并后插入一页之中。  
## 旋转图形的尺寸变化
通常图形的深度为0，总高度等于高度，但旋转后，其高度、宽度和深度都发生了变化。  
[图形旋转与高宽深的变化][2]  
图形顺时针旋转到90°时，高度为0，此时插图命令中设置任何高度值都将被提示出错。  
## 任意对象的旋转和缩放
`\includegraphics`只能对图形进行旋转或缩放，graphicx宏包另外提供了5条命令，可对任意LaTeX对象进行旋转或缩放。  
旋转命令：`\rotatebox[参数1=选项，参数2=选项，...]{角度}{对象}`  
### 旋转宏
rotating宏包利用`\rotatebox`等命令定义了3个可以旋转任意对象的旋转环境。  
`\begin{sideways} 对象 \end{sideways}` `\begin{turn}{角度} 对象 \end{turn}` `\begin{rotate}{角度} 对象 \end{rotate}`。  
第一个环境可将对象逆时针旋转90°，第二个和第三个都可将对象旋转任意角度，后者将对象视为一个空格，无论如何旋转，都不会改变对象的高度和宽度，这样对象很可能会与上下文重叠，但也可利用此特性制作各种文字特效。  
如果使用LaTeX编译源文件，大多数DVI文件阅读器不支持文本旋转，只有将其转换为PDF文件才能在PDF文件阅读器中正确显示。  
### 缩放对象
两种缩放命令。  
第一种设定缩放系数`\scalebox{水平缩放系数}[垂直缩放系数]{对象}`，垂直缩放系数默认等于水平缩放系数，水平缩放系数为负值则表示将对象左右反转180°，垂直缩放系数为负值，表示将对象上下颠倒。  
第二种是直接设定缩放对象的外形尺寸，还有带星号的形式，区别在于高度的控制，前者控制插图的高度，后者是总高度。  
`\resizebox{宽度}{高度}{对象}` `\resizebox*{宽度}{总高度}{对象}`。如果高度或宽度用`!`代替，表示将对象按照高度或者宽度并保持原高宽比进行缩放。可以使用尺寸测量命令`\height \width \totalheight`来表示对象的原始外形尺寸。  
### 镜像对象
`\reflectbox{对象}`将对象左右翻转180°，产生镜像效果。  
## 编译程序与图片格式
通常使用PDFLaTeX或者LaTeX编译程序，前者支持PNG、PDF和JPG格式，后者只支持EPS和PS格式。插图命令可以自动识别这两种编译程序，如果在插图命令的图形名中没有给出插图的扩展名，会自动寻找支持的格式并在找到后加上扩展名。  
通常应该在插图命令中给出图形文件的扩展名，以便缩短搜索图形文件的时间，避免内存空耗。如果插图命令没有找到适合的插图，系统将会终止编译并给出错误信息：`cannot find image file`。  
如果希望两种编译程序都能够使用，那么同一副插图应该有EPS和PDF（或PNG或JPG）两种格式的图形文件，并在插图命令中不给出扩展名，由其自行判断。  
## 图形格式的转换
使用LaTeX编译，必须全部使用EPS图形；如果使用PDFLaTeX或XeLaTeX编译，图形应为JPG、PDF或PNG。可以使用绘图软件另存为，也可以使用图形处理软件进行格式转换。  
### 使用绘图软件转换
Visio+Acrobat组合转换的图片格式尺寸较小，失真度较低。  
### 使用图形处理软件转换
绘图软件转换效果不佳或者转换后的尺寸过大。采用图形处理软件转换图形麻烦些，但是转换质量高，文件尺寸小。  
ImageMagick：跨平台，最主要功能是转换图形格式，是一个命令行工具。命令格式为:`convert 输入图形名 输出图形名`。可以使用通配符一次性转换多个文件。在Windows上可以写一个脚本方便重复使用。  
bmeps: CTex已附带。命令行格式`bmeps 选项 输入图姓名 输出图形名`。  
## 多图并列
如有多个外形尺寸较小的插图，可利用小页环境。`\begin{minipage}[位置]{宽度} end{minipage}`嵌套在figure环境中使用。  
## 插图的边框
宏包xcolor提供的彩色边框命令可为插图添加可设置颜色的边框，即`\fcolorbox`。  
## 图文绕排
通常，插图都是在两段文本之间，单独一行，水平居中，左右两侧留有相同宽度的空白。但有时图形很小，比如徽标或某人照片，如果也独处一行，两侧的空白区域就远大于插图幅面，看起来很不协调。最好是图文绕排。  
## 绕排宏包pincinpar
该宏包提供了3种用于图文绕排的环境。  
1. `\begin{window}[行数，位置{绕排对象}，{标题}] 绕排文本 \end{window}`。  
2. `\begin{figwindow}[行数，位置{绕排对象}，{标题}] 绕排文本 \end{figwindow}`。  
3. `\begin{tabwindow}[行数，位置{绕排对象}，{标题}] 绕排文本 \end{tabwindow}`。  

行数：指定绕排于绕排对象上方的文本行数。  
位置：设定绕排对象在版面的位置，用l、c、r分别表示左、中、右。  
绕排对象：可以是图形、表格或文本。  
标题：绕排对象的标题。  
上述三种环境功能标题，只是后两种可以在绕排对象的标题前自动添加标题标志和分隔符号。  
## 其它绕排宏包
wrapfig：提供wrapfigure和wraptable两环境，可在小页环境中对图表进行绕排。  
floatflt：提供floatingfigure和floatingtable环境，可按左右页确定绕排位置。
